```python
df['Precip Type'] = df['Precip Type'].fillna(df['Precip Type'].mode()[0])
```

```python
df['Temperature (C)'].max()
```

```python
df = df.drop(['Formatted Date','Summary','Daily Summary','Wind Bearing (degrees)'],axis =1)
```

```python
X = df.drop(['Temperature (C)','Loud Cover'],axis =1 )
Y = df['Temperature (C)']
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
lbl = LabelEncoder()
X['Precip Type'] = lbl.fit_transform(X['Precip Type'])
```

```python
X
```

```python
plt.figure(figsize = (15,15))
crr = X.corr()
sns.heatmap(crr,annot = True)
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test = train_test_split(X,Y,test_size = .2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_test.isnull().sum()
```

```python
y_p = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error , r2_score
mean_squared_error(y_p,y_test)
```

```python
mean_absolute_error(y_p,y_test)
```

```python
r2_score(y_p,y_test)
```
